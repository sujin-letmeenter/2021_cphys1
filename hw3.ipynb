```python
def factorial(n):
    if n==1:
        return 1
    return n*factorial(n-1)

def f(x,N):
    i,s=1,0
    while(i<=N):
        s=s+(-1)**i*x**(2*i)/factorial(2*i-1)
        i+=1
    return s

x=np.arange(-5,5,0.1)
print(x,f(x,1))
plt.plot(x,f(x,1))
```
