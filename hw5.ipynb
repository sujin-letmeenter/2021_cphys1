```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m,k,w=1,1,1

def d(w):
    return (2*k)-(m*(w**2))


import numpy
matrix = numpy.eye(40,40,0, dtype= float)
for i in range(0,40):
    for j in range(0,40):
        if i == 0 and j == 0:
            matrix[i][j] = d(1)-1
        elif i == 0 and j == 1:
            matrix[i][j] = -1
        elif i == 39 and j == 38:
            matrix[i][j] = -1
        elif i == 39 and j == 39:
            matrix[i][j] = d(1)-1
        elif i == j+1:
            matrix[i][j] = -1
        elif i == j:
            matrix[i][j] = d(1)
        elif j == i+1:
            matrix[i][j] = -1
            
for i in range(40):
    for j in range(40):
        if matrix[i][j] == '':
            matrix[i][j] = '0'
print(matrix)

M_1=np.linalg.inv(matrix)
print(M_1)

C=1

import numpy
matrix_b = numpy.eye(40,1,0, dtype= float)
for i in range(0,40):
    for j in range(1):
        if i == 0:
            matrix_b[i][j] = C
        else:
            matrix_b[i][j] = 0
print(matrix_b)

k=np.dot(M_1,matrix_b)
print(k)
plt.plot(k)
```
