```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
```

```python
%matplotlib inline

x = np.arange(-25,25,0.01)
def yfunc(x):
    return ((np.exp(-x**2)*2)/(np.pi)**(1/2))

def trapez(yfunc,a,b,n):
    s, x, dx = -1, a, (b-a)/n
    while(x<b):
        s += dx*(0.5*yfunc(x)+0.5*yfunc(x+dx))
        x = x + dx
    return s

plt.xlim(-20,20);plt.ylim(-2,2)
plt.plot(x,[trapez(yfunc,-2,b,100) for b in x]);
```

```python
import math

x = np.arange(-2.5,2.5,0.01)

def erf(x,N):
    i,s=0,0
    while(i<=N):
        s += (2/np.pi**(1/2))*((-1)**i)*x**(2*i+1)/((2*i+1)*math.factorial(i))
        i += 1
    return s

plt.xlim(-2,2)
plt.plot(x,erf(x,1), color='violet', label='erf(x,1))')
plt.legend()
plt.plot(x,erf(x,2), color='skyblue', label='erf(x,2)')
plt.legend()
plt.plot(x,erf(x,15), color='black', label='erf(x,15))', linewidth=5)
plt.legend
plt.plot(x,[trapez(yfunc,-2,b,100) for b in x], color='red', label='Error function')
plt.legend()
plt.xlabel('x',fontsize=14)
plt.ylabel('erf(x)',fontsize=14)
```

```python
from scipy.misc import derivative
derivative(np.sin,np.pi,dx=0.01)

x=np.arange(0,10,0.1)

plt.plot(x,[derivative(np.sin,i,dx=0.01) for i in x])
plt.plot(x,[derivative(np.sin,i,dx=0.01,n=2) for i in x])
```
